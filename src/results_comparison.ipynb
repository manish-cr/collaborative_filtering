```python
import matplotlib.pyplot as plt

# RMSE values
item_rmse_naive = 3.69
item_rmse_surprise_knn = 1.09
item_rmse_fastai = 0.52

# Labels for bars
labels = ['Naive', 'Surprise KNN', 'FastAI (latent matrix factorization)']

# RMSE values
rmse_values = [item_rmse_naive, item_rmse_surprise_knn, item_rmse_fastai]

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(labels, rmse_values, color=['blue', 'green', 'red'])
plt.xlabel('Methods')
plt.ylabel('RMSE')
plt.title('RMSE Comparison')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
